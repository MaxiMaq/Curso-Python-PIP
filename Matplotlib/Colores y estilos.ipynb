```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig, ax=plt.subplots(figsize=(8,8)) 

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'gx:')
```

Pyplot

```python
fig, ax = plt.subplots(figsize = (8,8))

ax.plot(x,x+1,color = '#D426C8', alpha=0.8, linewidth=5, linestyle='-', marker='o', markersize='10', markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color = '#66FF89', linewidth=8, linestyle='--', marker='x')
ax.plot(x,x+3,color = 'blue', linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4, color = 'black', linewidth=2, linestyle=':', marker='P')
```
